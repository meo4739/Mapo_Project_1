```python
def question_analysis(origin_df, question, year, origin_value, replace_value):
    result_df = pd.DataFrame()
    sample_df = pd.DataFrame()
    result_df = origin_df.loc[:, question].replace({origin_value[0] : replace_value[0],
                                                    origin_value[1] : replace_value[1],
                                                    origin_value[2] : replace_value[2],
                                                    origin_value[3] : replace_value[3]} ).value_counts().to_frame().reset_index()
    
    sampleNum_df = pd.DataFrame({'count' : [ origin_df.loc[origin_df[question].isna()].shape[0],
                                        origin_df.loc[~origin_df[question].isna()].shape[0],
                                        origin_df.shape[0]]})
    sampleNum_df.index = ['응답자', '미응답자', '전체표본']
    sampleNum_df = sampleNum_df.reset_index().rename({'index' : question}, axis=1)
    sampleNum_df.index = [4,5,6]
    result_df = pd.concat([result_df, sampleNum_df], axis=0)
    result_df.rename({'count':year}, inplace=True, axis=1)

    return result_df

df_2020_Q14 = question_analysis(df_2020, '문14. 현재 살고 계신 임대주택은 어떤 유형입니까?_1) 유형', '2020', [1.0, 2.0, 3.0, 4.0], ['민간임대주택', '공공임대주택', '분양전환공공임대주택', '기타임대주택'])
df_2021_Q14 = question_analysis(df_2021, '문14. 현재 살고 계신 임대주택은 어떤 유형입니까?_1) 유형', '2021', [1.0, 2.0, 3.0, 4.0], ['민간임대주택', '공공임대주택', '분양전환공공임대주택', '기타임대주택'])
df_2022_Q14 = question_analysis(df_2022, '문14. 현재 살고 계신 임대주택은 어떤 유형입니까?_1) 유형', '2022', [1.0, 2.0, 3.0, 4.0], ['민간임대주택', '공공임대주택', '분양전환공공임대주택', '기타임대주택'])

df_22_24_Q14 = pd.concat([df_2020_Q14, df_2021_Q14, df_2022_Q14], axis=1)
df_22_24_Q14 = df_22_24_Q14.loc[:, ~df_22_24_Q14.columns.duplicated()]

display(df_22_24_Q14)


def ratio_calculate(origin_df, list_all):
    ratio_df = origin_df.copy()
    for index, i in enumerate(list_all):
        origin_df[i] = origin_df[i].astype(float)
        ratio_df[i] = round(origin_df.iloc[:,index+1] / origin_df.iloc[-1:,index+1].values * 100,2)
    return ratio_df

df_22_24_Q14_ratio = ratio_calculate(df_22_24_Q14, ['2020', '2021', '2022'])
display(df_22_24_Q14_ratio)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def multi_barGraph(graph_num, graph_size, df, target, col_list, title_list):
    
    fig, axes = plt.subplots(graph_num[0], graph_num[1], figsize=graph_size)

    for i in range(len(col_list)):
        sns.barplot(x=df[target][:-3].values, y=df[col_list[i]][:-3].values, ax=axes[i])
        axes[i].set_title(title_list[i])
        axes[i].tick_params(axis='x', rotation=45)
        
        for bar in axes[i].patches:
            height = bar.get_height()
            axes[i].annotate(f'{height:.2f}', 
                             xy=(bar.get_x() + bar.get_width() / 2, height), 
                             xytext=(0, 3), 
                             textcoords="offset points", 
                             ha='center', va='bottom')
    plt.tight_layout()
    plt.show()
    
multi_barGraph((1,3), (14,7), df_22_24_Q14, 
               '문14. 현재 살고 계신 임대주택은 어떤 유형입니까?_1) 유형',
               ['2020','2021', '2022'],
               ['2020 임대주택 유형 숫자', '2021 임대주택 유형 숫자', '2022 임대주택 유형 숫자'])

multi_barGraph((1,3), (14,7), df_22_24_Q14_ratio, 
               '문14. 현재 살고 계신 임대주택은 어떤 유형입니까?_1) 유형',
               ['2020','2021', '2022'],
               ['2020 임대주택 유형 비율', '2021 임대주택 유형 비율', '2022 임대주택 유형 비율'])
```
